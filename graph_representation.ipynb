```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# 2.	Using sets and dictionary libraries represent the graph in Figure 1 into a python graph structure.
# Creating an empty graph using DiGraph
G = nx.DiGraph()

# Defining the nodes and heuristics
nodes = ['S', 'A', 'B', 'C', 'D', 'G']
heuristics = {
    'S': 7,
    'A': 5,
    'B': 7,
    'C': 4,
    'D': 1,
    'G': 0
}

# Adding nodes with heuristics as labels
for node in nodes:
    G.add_node(node, heuristic=heuristics[node])

# Defining edges and their weights
edges = [('S', 'A', 3), ('S', 'B', 1), ('A', 'B', 2), ('A', 'C', 2),
         ('B', 'C', 3), ('C', 'D', 4), ('C', 'G', 4), ('D', 'G', 1)]

# Adding directed edges with weights
for edge in edges:
    G.add_edge(edge[0], edge[1], weight=edge[2])

# Defining node positions for better layout
node_positions = {
    'S': (0, 2),
    'A': (1, 3),
    'B': (1, 1),
    'C': (2, 2),
    'D': (3, 3),
    'G': (3, 1),
}

# Drawing the graph
edge_labels = {edge: G.get_edge_data(*edge)['weight'] for edge in G.edges()}

# Creating labels for the nodes with heuristics
node_labels = {node: f"{node}\nH={heuristics[node]}" for node in nodes}

# Drawing the nodes and edges
nx.draw(G, node_positions, with_labels=False, node_size=500, node_color='skyblue')
nx.draw_networkx_edge_labels(G, node_positions, edge_labels=edge_labels, font_size=10, font_color='red')
nx.draw_networkx_labels(G, node_positions, labels=node_labels, font_size=10, font_color='black')


plt.title("Graph Representation")
plt.show()
```
